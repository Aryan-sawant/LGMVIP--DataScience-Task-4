**Image to Pencil Sketch**

**Importing necessary required Libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
```

**Reading the Image**

```python
img=cv2.imread('/content/Shri Krishna Supreme Universal Form.jpg')
```

**Displaying image using OpenCV**

```python
cv2_imshow(img)
```

**Converting to Gray Image**

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

**Inverting the Image**

```python
invert_img=cv2.bitwise_not(gray_img)
cv2_imshow(invert_img)
```

**Blurring the Image**

```python
blur_img=cv2.GaussianBlur(invert_img,(111,111),0)
cv2_imshow(blur_img)
```

**Inverting the Blurred Image**

```python
invert_blur_img=cv2.bitwise_not(blur_img)
cv2_imshow(invert_blur_img)
```

**Sketch**

```python
sketch_img=cv2.divide(gray_img,invert_blur_img,scale=256)
```

**Saving the Sketch**

```python
cv2.imwrite('sketch.png',sketch_img)
```

**Displaying the Sketch Image**

```python
cv2_imshow(sketch_img)
```
